```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/refs/heads/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
#aggregate by date
df = df.groupby(["Country/Region"]).sum().drop(columns=["Lat","Long"]).T
df["US"].plot(label="USA",figsize=(10,5))
df["India"].plot(label="India")
plt.legend()
plt.title("COVID-19 Cases Over Time")
plt.show()
```
